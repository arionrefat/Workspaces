```python
#Task 1
class Student:
    ID=0
    def __init__(self, name, department, age, cgpa):
        self.name = name
        self.department = department
        self.age = age
        self.cgpa = cgpa

    def get_details(self):
        Student.ID += 1
        print(f'ID: {Student.ID}')
        print(f'Name: {self.name}')
        print(f'Department: {self.department}')    
        print(f'Age: {self.age}')
        print(f'CGPA: {self.cgpa}')
    
    @classmethod
    def from_String(cls,data):
        name,department,age,cgpa=data.split('-')
        obj=cls(name, department, age, cgpa)
        return obj

s1 = Student("Samin","CSE",21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE",21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura","EEE",22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

print('\nAnswer to subtask 5:\nA class variable is a variable that can be used and accessed throughout the class. On the otherhand, an instance variable is a variable which can only be accessed within the instance method where it is defined.')
print('\nAnswer to subtask 6:\nAn instance method is a method within a class that takes attributes from the object as arguments/parameters and is able to freely modify the attributes of the object and class. On the other hand, Class method is a method that takes the parameters of the class as arguments/parameters and is able to modify the attributes of the class as a whole. It cannot modify object specific attributes which Instance method can.')
```

```python
#Task 3
class Dates:
    def __init__(self,date):
        self.date = date
    @classmethod
    def toDashDate(cls, date):
        d,m,y=date.split('/')
        date = d +'-' + m + '-' + y
        date=cls(date)
        return date
    def getDate(self):
        return self.date

date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate()):
    print("Equal")
else:
    print("Unequal")

print("\nThe output is equal because the returned value of the date parameter of the objects 1 and 2 are same. We used the modified class method to fix the form of the date parameter of second object which is the same as the form of the data parameter of the first class. Thus, the output from the conditional statement is true, which makes an output of 'Equal'")
```
