```python
#Easy - Task 4

num = int(input("Enter your number: "))
  
if num > 1:   
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
        else:
            print(num, "is a prime number") 
else:
    print(num, "is not a prime number") 
```

```python
#Hard - Task 3

number = int(input("Enter your number: "))

sum = 0

for i in range(1,number):
    if number%i==0:
        sum = sum+i
if(sum == number):
    print("Perfect number")
else:
    print("Not perfect")
```

```python
#Hard - Task 1

number = input("Enter your numnber: ")
size = len(number)
count = 0

for i in range(size):
    temp = number[i]
    for j in range(size-1):
        if temp != number[j]:
            count = count + 1
print(count)
```

```python
#Hard - Task 2

sizeN = int(input("Enter rows: "))

for countRow in range(1, sizeN+1):
    for countSpace in range(1, (sizeN - countRow)+1):
        print(" ",end="")
        
    for countStars in range(1, 2*countRow):
        if(countStars ==1 or countStars==2*countRow - 1 or countRow == sizeN ):
            print("*",end="")
        else:
            print(" ",end="")
    print("\n")
    
for countRow in range(sizeN+1, 2*sizeN):
    for countSpace in range(1, countRow - sizeN+1):
        print(" ",end="")
    
    for countStars in range(1, 4*sizeN-1-2*countRow+1):
        if(countStars == 1 or countStars == 4*sizeN-1-2*countRow):
            print("*",end="")
        else:
            print(" ",end="")
    print("\n")
        
```
