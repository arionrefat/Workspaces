```python
#Question - 1

class Calculator:
    def __init__(self,num1,sign,num2):
        self.sign = sign
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return num1 + num2
    def substract(self):
        return num1 - num2
    def divide(self):
        return num1//num2
    def multiplication(self):
        return num1*num2

num1 = int(input())
sign = input()
num2 = int(input())

calculate = Calculator(num1,sign,num2)
print("Let's Calculate!")
if calculate.sign == "+":
    result = calculate.add()
if calculate.sign == "-":
    result = calculate.add()
if calculate.sign == "*":
    result = calculate.add()
if calculate.sign == "/":
    result = calculate.add()
print("Value 1:",num1,"\n"
      "Operator:", sign, "\n"
      "Value 2:", num2, "\n"
      "Result:", result, "\n")
```

```python
#Question - 2

class Customer:
    def __init__(self,name):
        self.name = name
    def greet(self,name = None):
        if name != None:
            print(f"Hello {name}!")
        else:
            print("Hello!")
    def purchase(self,*args):
        count=0
        for arg in args:
            count+=1
        print(f'{self.name}, you have purchased {count} item(s):')
        for i in args:
            print(i)
        
customer_1 = Customer("Sam")
customer_1.greet("David")
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
        
```

```python
#Question - 3

class Panda:
    def __init__(self,name,gender,age):
        self.age = age 
        self.gender = gender
        self.name = name
    def sleep(self,time = None):     
        if time != None and 3 <= time <= 5:
            return (f"{self.name} sleeps {time} hours daily and should have Mixed Veggies")
        elif time != None and 6 <= time <= 8:
            return (f"{self.name} sleeps {time} hours daily and should have Eggplant & Tofu")
        elif time != None and 9 <= time <= 11:
            return (f"{self.name} sleeps {time} hours daily and should have Broccoli")
        else:
            return (f"{self.name} sleeps {time} hours daily and should have only bamboo leaves")

panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} yearsold".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} yearsold".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} yearsold".format(panda3.name,panda3.gender,panda3.age))

print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#Question - 4

class Cat:
    def __init__(self,color = "White",stance = "sitting"):
        self.color = color 
        self.stance = stance
    def printCat(self):
        print(f"{self.color} cat is {self.stance}")
    def changeColor(self,color):
        self.color=color
c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
#Question - 5

class Vehicle:
    def __init__(self,x = 0, y = 0): 
        self.x = x 
        self.y = y
    def print_position(self):
        print(f"({self.x},{self.y})")
    def moveUp(self):
        self.y = self.y + 1
    def moveLeft(self):
        self.x = self.x - 1
    def moveDown(self):
        self.y = self.y - 1
    def moveRight(self):
        self.x = self.x + 1
        
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```
